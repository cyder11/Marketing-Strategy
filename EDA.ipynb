```python
bb= []
for counter, value in enumerate(b):
    if 'Leisure' in b[counter]:
        bb.append('Leisure')
    elif 'Business' in b[counter]:
        bb.append('Business')
    else:
        bb.append('False')
```

```python
cc= []
for counter, value in enumerate(b):
    if 'Solo' in b[counter]:
        cc.append('Solo')
    elif 'Family' in b[counter]:
        cc.append('Familiy')
    elif 'Couple' in b[counter]:
        cc.append('Couple')
    elif 'Group' in b[counter]:
        cc.append('Group')
    else:
        cc.append('False')
```

```python
oo=[]
import re
for counter, value in enumerate(o):
    oo.append(re.findall(r'\d+', o[counter]))
```

```python
qq=[]
for e in oo:
    qq.append(int(e[0]))
    
```

```python
ee= []
for counter, value in enumerate(b):
    if '1' in b[counter]:
        ee.append('1')
    elif '2' in b[counter]:
        ee.append('2')
    elif '3' in b[counter]:
        ee.append('3')
    elif '4' in b[counter]:
        ee.append('4')
    elif '5' in b[counter]:
        ee.append('5')
    elif '6' in b[counter]:
        ee.append('6')
    elif '7' in b[counter]:
        ee.append('7')
    elif '8' in b[counter]:
        ee.append('8')
    elif '9' in b[counter]:
        ee.append('9')
    else:
        ee.append('+15')
```

```python
ff= []
for counter, value in enumerate(b):
    if '1' in b[counter]:
        ff.append(1)
    elif '2' in b[counter]:
        ff.append(2)
    elif '3' in b[counter]:
        ff.append(3)
    elif '4' in b[counter]:
        ff.append(4)
    elif '5' in b[counter]:
        ff.append(5)
    elif '6' in b[counter]:
        ff.append(6)
    elif '7' in b[counter]:
        ff.append(7)
    elif '8' in b[counter]:
        ff.append(8)
    elif '9' in b[counter]:
        ff.append(9)
    else:
        ff.append(0)
```
